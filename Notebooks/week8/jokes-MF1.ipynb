<p>This document is about recommender systems. You are trying to learn P and Q that help predict the actual rating. . </p>

```python
# we are trying to learn these factors especially P and Q that are going to be able to reproduce our original matrix. 
# You start with something random like we said and you have to greadient discent and try to get towards zero. 
# you give it how much you would like to truncate those factors (k) and that is something you might wana tune for
# This function is about learning your matrix factors using gradient discent
def matrix_factorization(R, P, Q, K, steps=5000, alpha=0.0002, beta=0.02):
    Q = Q.T
    for step in range(steps):
        for i in range(len(R)): # go through out data
            for j in range(len(R[i])): #
                if R[i][j] > 0: # look for something that is positive
                        # error is equal to the actual value the person has given in their training data minus our prediction using p and q factors
                    eij = R[i][j] - np.dot(P[i,:],Q[:,j]) #2:40 
                    for k in range(K):
                        P[i][k] = P[i][k] + alpha * (2 * eij * Q[k][j] - beta * P[i][k])
                        Q[k][j] = Q[k][j] + alpha * (2 * eij * P[i][k] - beta * Q[k][j])
        eR = np.dot(P,Q)
        e = 0
        for i in range(len(R)):
            for j in range(len(R[i])):
                if R[i][j] > 0:
                    e = e + pow(R[i][j] - np.dot(P[i,:],Q[:,j]), 2)
                    for k in range(K):
                        e = e + (beta/2) * ( pow(P[i][k],2) + pow(Q[k][j],2) )
        if e < 0.001: # if the error is smaller, just stop
            break
        print ("Step %d of %d; Error: %0.5f; Time: %0.2f" %(step+1, steps, e, time()))
    return P, Q.T
```

```python
M = len(dataMat)
N = len(dataMat[0])
K = 5
steps = 3000
# 
```

```python
P = np.random.rand(M,K)
Q = np.random.rand(N,K)
```

```python
from time import time
t0 = time()
fP, fQ = matrix_factorization(dataMat, P, Q, K, steps=steps)
print("done in %0.3fs." % (time() - t0))
```

```python
outP = open("../../data/jokes_p.csv", "w")
outQ = open("../../data/jokes_q.csv", "w")
np.savetxt(outP, fP, delimiter=',', fmt='%1.4f')
np.savetxt(outQ, fQ, delimiter=',', fmt='%1.4f')
```

<p>This is not SVD but this is trying to learn so that P * Q gives you your original matrix .  </p>

```python
# You can an erro per user instead of the total error
# You get an error of every single user in your matrix
totCount = 0
totError = 0
for u in range(M):
    err_u = 0
    rateCount_u = 0
    for j in range(N):
        if (dataMat[u,j] > 0):
            rateCount_u += 1
            err_u += abs(np.dot(fP[u],fQ[j]) - dataMat[u,j])
    print ("Mean Absolute Error for User %d = %0.3f" %(u, err_u/rateCount_u))
    totCount += rateCount_u
    totError += err_u

print()
print ("Overall Mean Absolute Error = %0.3f" %(totError/totCount))
```
